1. Use el m茅todo de bisecci贸n para encontrar soluciones precisas dentro de  $ 10^{-2}$ para $ x^{3} -7x^{2} + 14x-6=0$  en cada intervalo.

    a. $ [0,1] $
    b. $[1, 3.2]$
    c. $[3.2, 4]$ 

```python

# Definici贸n de la funci贸n
def f(x):
    return x**3 - 7*x**2 + 14*x - 6


# M茅todo de bisecci贸n
def biseccion(a, b, tol=1e-2, max_iter=50):
    print(f"\nIntervalo inicial: [{a}, {b}]")
    
    if f(a) * f(b) >= 0:
        print("El m茅todo de bisecci贸n no se puede aplicar: f(a) y f(b) deben tener signos opuestos.")
        return None

    iteracion = 0
    while abs(b - a) > tol and iteracion < max_iter:
        c = (a + b) / 2.0
        fc = f(c)

        if fc == 0:
            return c  # Se encontr贸 la ra铆z exacta

        if f(a) * fc < 0:
            b = c
        else:
            a = c

        iteracion += 1
        print(f"Iteraci贸n {iteracion}: c = {c:.6f}, f(c) = {fc:.6f}, rango = [{a:.6f}, {b:.6f}]")

    return (a + b) / 2.0


# B煤squeda de ra铆ces en diferentes intervalos
print("\n=== B煤squeda en [0, 1] ===")
raiz = biseccion(0, 1)
print(f"Ra铆z aproximada: {raiz:.6f}")

print("\n=== B煤squeda en [1, 3.2] ===")
raiz = biseccion(1, 3.2)
print(f"Ra铆z aproximada: {raiz:.6f}")

print("\n=== B煤squeda en [3.2, 4] ===")
raiz = biseccion(3.2, 4)
print(f"Ra铆z aproximada: {raiz:.6f}")
```
